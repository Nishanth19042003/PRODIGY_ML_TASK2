```python
import pandas as pd
```

```python
customer_data = pd.read_csv(r"C:\Users\NISHANTH\Documents\sem 8\intern\Mall_Customers.csv")
customer_data
```

```python
# Display basic information
customer_data.info()

# Preview the data
customer_data.head()

# Check for missing values
print(customer_data.isnull().sum())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Select relevant features for clustering (e.g., Annual Income and Spending Score)
features = ['Annual Income (k$)', 'Spending Score (1-100)']
X = customer_data[features]

# Test different numbers of clusters
inertias = []
cluster_range = range(1, 11)  # Try 1 to 10 clusters

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the elbow graph
plt.plot(cluster_range, inertias, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Method for Optimal Clusters')
plt.show()
```

```python
# Based on the elbow graph, choose the optimal number of clusters (e.g., 5)
optimal_clusters = 5
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)

# Fit K-means to the data
kmeans.fit(X)

# Add the cluster labels to the original data
customer_data['Cluster'] = kmeans.labels_
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green', 'purple', 'orange']  # Adjust colors as needed

for i in range(optimal_clusters):
    cluster_data = customer_data[customer_data['Cluster'] == i]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], 
                s=100, c=colors[i], label=f'Cluster {i}')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Clusters')
plt.legend()
plt.show()
```

```python
# Summary of clusters
cluster_summary = customer_data.groupby('Cluster').mean()

print(cluster_summary)
```
